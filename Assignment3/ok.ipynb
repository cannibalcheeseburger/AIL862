```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['500 Samples', '1000 Samples', 'Domain Adaptation']
train_acc = [99.17, 99.19, 98.79]
val_acc = [99.33, 99.58, 99.63]
test_acc = [84.55, 87.43, 91.62]

# Bar positions and width
x = np.arange(len(methods))  # the label locations
width = 0.25  # the width of the bars

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars with improved colors
rects1 = ax.bar(x - width, train_acc, width, label='Train Acc', color='#1f77b4', alpha=0.85)
rects2 = ax.bar(x, val_acc, width, label='Val Acc', color='#2ca02c', alpha=0.85)
rects3 = ax.bar(x + width, test_acc, width, label='Test Acc', color='#d62728', alpha=0.85)

# Add labels, title, and legend
ax.set_ylabel('Accuracy (%)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(methods, fontsize=11)
ax.legend(fontsize=10)

# Function to add value labels on top of each bar
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}%',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 5),  # Adjusted vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10)

# Apply the function to all bars
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# Adjust layout to prevent label clipping
fig.tight_layout(pad=2)

# Show plot
plt.show()
```
